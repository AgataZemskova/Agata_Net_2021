Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из России - ставим other
все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log.loc[log['traffic_source'] == 'google']
```

```python
def source_type(row):
    if ('yandex' or 'google') in row['traffic_source']:
        return 'organic'
    elif ('paid' or 'email') in row['traffic_source'] and 'Russia' in row['region']:
        return 'ad'
    elif ('paid' or 'email')in row['traffic_source'] and 'Russia' not in row['region']:
        return 'other'
    else:
        return row['traffic_source']
    
```

```python
log['source_type'] = log.apply(source_type, axis=1)
log
```

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
r = pd.read_csv('ratings.csv')
r.head()
```

```python
r_grouped = r.groupby('userId', as_index=False).count()
```

```python
r_grouped = r_grouped.loc[r_grouped.rating > 100, :]
r_grouped
```

```python
r_grouped_max = r.groupby(['userId'], sort=False)['timestamp'].max()
```

```python
r_grouped_min = r.groupby(['userId'], sort=False)['timestamp'].min()
```

```python
r_grouped = r_grouped.merge(r_grouped_max, on='userId').merge(r_grouped_min, on='userId')
```

```python
r_grouped['lifetime'] = r_grouped['timestamp_y'] - r_grouped['timestamp_x']
r_grouped[['userId', 'lifetime']]
```
