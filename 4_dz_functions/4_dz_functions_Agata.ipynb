Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

Задание 1
Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

```python
def cmd_input():
    cmd = 'Waiting'
    while cmd != 'q':
        cmd = input('Введите команду: ')
        if cmd == 'p':
            doc_number()
        elif cmd == 's':
            directory()
        elif cmd == 'l':
            all_info()
        elif cmd == 'ads':
            add_shelf()
        elif cmd == 'ds':
            del_shelf()
        else:
            break


"""
Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
"""
def doc_number():    
    doc = input('Введите номер документа: ') 
    for i in documents:
        flag = 0
        if doc == i['number']:
            get_owner = i.get('name')
            print('Владелец документа: ' + get_owner)
            flag = 1
            break
    if flag == 0:
        print('Документ не найден в базе')
    return('ok')
  

"""
Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
"""
def directory():
    dir = input('Введите номер документа: ') 
    for i in directories:
        flag = 0
        if dir in directories[i]:
            print('Документ хранится на полке: ' + i)
            flag = 1
            break
    if flag == 0:                         
        print('Документ не найден в базе')
    return('ok')
 

"""
Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
"""
def all_info():
    count = 0
    for i in documents:
        if count < len(documents):
            for j in directories:
                if i['number'] in directories[j]:
                    shelf = j
                    print('№ ' + i['number'] + ', тип: ' + i['type'] + ', владелец: ' + i['name'] + ', полка хранения: ' + shelf)
            count +=1  
    return('ok')       
            
  
"""
Пункт 4. Пользователь по команде "ads" может добавить новую полку
"""
def add_shelf():
    num_shelf = input('Введите номер полки: ') 
    if num_shelf in directories:
        dir_list = ', '.join(directories)
        print('Такая полка уже существует. Текущий перечень полок:  ' + str(dir_list))
    else:
        directories[num_shelf] = '[]'
        dir_list = ', '.join(directories)
        print('Полка добавлена. Текущий перечень полок:  ' + str(dir_list))
    return('ok') 


"""
Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
"""
def del_shelf():
    num_shelf = input('Введите номер полки: ') 
    if num_shelf in directories:
        if directories[num_shelf] == '[]':
            del directories[num_shelf]
            dir_list = ', '.join(directories)
            print('Полка удалена. Текущий перечень полок: ' + str(dir_list))
        else:
            dir_list = ', '.join(directories)
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + str(dir_list))            
    else:
        dir_list = ', '.join(directories)
        print('Такой полки не существует. Текущий перечень полок: ' + str(dir_list))
    return('ok')   



cmd_input()        
        
```
