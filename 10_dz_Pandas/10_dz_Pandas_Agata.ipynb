Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
df_r = movies.merge(ratings, on='movieId')
df_r.head()
```

```python
df_r[df_r.rating == 5.0].groupby('title').count().sort_values('rating', ascending=False).head(1)
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
df = pd.read_csv('power.csv')
df.head()
```

```python
baltic = df[ ( (df['country'] == 'Latvia') | (df['country'] == 'Lithuania') | (df['country'] == 'Estonia') ) 
            & ( (df['category'] == 4)  | (df['category'] == 12) | (df['category'] == 21) ) 
            & ( (df['year'] <= 2010) & (df['year'] >= 2005) )
            & (df['quantity'] > 0) ]
baltic['quantity'].sum()
```
