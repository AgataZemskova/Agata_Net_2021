Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
df = movies.merge(ratings, on='movieId')
df.head()
```

```python
def rating_f(row):
    if 0.5 <= row <= 2:
        return 'низкий рейтинг'
    elif 4.5 <= row <= 5:
        return 'высокий рейтинг'
    else:
        return 'средний рейтинг'
```

```python
df['rating_class'] = df['rating'].apply(rating_f)
df.head()
```

```python
kw = pd.read_csv('keywords.csv', encoding='UTF-8')
kw.head(100)
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
for k in kw.keyword:
    if k in (list(item for sublist in geo_data.values() for item in sublist)):
        for reg, val in geo_data.items():
            if k in val:
                print(k, reg)
            else:
                print('undefined') 
```

```python
def region2(row):
    row = row.split(' ')
    for k in row:
        if k in (list(item for sublist in geo_data.values() for item in sublist)):
            for reg, val in geo_data.items():
                if k in val:
                    return reg
                else:
                    return('undefined') 
```

```python
kw['region'] = kw['keyword'].apply(region2)
kw[kw['keyword'].str.contains('псков')]
```
