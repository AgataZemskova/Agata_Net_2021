```python
#выбросы rectal_temperature (по тем, где значения 1 и 2, не будем искать)
q1_rectal_temperature = df['rectal_temperature'].quantile(0.25)
q3_rectal_temperature = df['rectal_temperature'].quantile(0.75)
iqr_rectal_temperature = q3_rectal_temperature - q1_rectal_temperature
lower_bound_rectal_temperature = q1_rectal_temperature - (1.5 * iqr_rectal_temperature) 
upper_bound_rectal_temperature = q3_rectal_temperature + (1.5 * iqr_rectal_temperature)
remove_outliers_rectal_temperature = df[df['rectal_temperature'].between(lower_bound_rectal_temperature, upper_bound_rectal_temperature, inclusive=True)].sort_values('rectal_temperature')
remove_outliers_rectal_temperature
```

```python
pd.concat([df, remove_outliers_rectal_temperature]).drop_duplicates(keep=False).dropna()
```

```python
#функция не получилась, нужно drop_duplicates делать вне ее? 
def outliers_search(col):
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = df[col.between(lower_bound, upper_bound, inclusive=True)].sort_values(col)
    remove_outliers
    return(pd.concat([df, remove_outliers]).drop_duplicates(keep=False).dropna())
    
```

```python
df['outliers_rectal_temperature'] = df.apply(outliers_search)
```

```python
df.dropna(subset=['surgery?'], inplace=True)
```

```python
df.isnull().sum(axis = 0)
```

```python
#замена медианой для rectal_temperature
df['rectal_temperature'].fillna(df.groupby('pain')['rectal_temperature'].transform('median'), inplace=True)
```

```python
#замена медианой для pulse
df['pulse'].fillna(df.groupby('pain')['pulse'].transform('median'), inplace=True)
```

```python
#замена медианой для respiratory_rate
df['respiratory_rate'].fillna(df.groupby('pain')['respiratory_rate'].transform('median'), inplace=True)
```

```python
#замена модой для temperature_of_extremities
df['temperature_of_extremities'].fillna(df['temperature_of_extremities'].mode(), inplace=True)
```

```python
#замена модой для pain
df['pain'].fillna(df['pain'].mode()[0], inplace=True)
```
